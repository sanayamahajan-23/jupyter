```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
iris_dataset=load_iris()
X,y=iris_dataset.data,iris_dataset.target
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
lr=LinearRegression().fit(X_train,y_train)
print("lr.coef_:{}",format(lr.coef_))
print("lr.intercept_:{}",format(lr.intercept_))
print("training score:{:.2f}",format(lr.score(X_train,y_train)))
print("test set score:{:.2f}",format(lr.score(X_test,y_test)))
import matplotlib.pyplot as plt
import numpy as np
y_pred=lr.predict(X_test)
plt.scatter(y_pred,y_test,color='blue',label='actual')
plt.xlabel("actual target value")
plt.ylabel("predicted target value")
plt.title("actual vs predicted")
plt.plot([0, 2], [0, 2], 'r--',label='predicted line')
plt.grid(True)
plt.legend()
plt.show()
X_single_feature = X[:, 2].reshape(-1, 1)
X_train_f, X_test_f, y_train_f, y_test_f = train_test_split(X_single_feature, y, random_state=42)

lr_single = LinearRegression().fit(X_train_f, y_train_f)

# Predict
y_pred_line = lr_single.predict(X_test_f)

# Plot
plt.scatter(X_test_f, y_test_f, color='green', label='Actual')
plt.plot(X_test_f, y_pred_line, color='black', linewidth=2, label='Regression line')
plt.xlabel("Petal Length")
plt.ylabel("Target")
plt.title("Linear Regression on One Feature")
plt.legend()
plt.grid(True)
plt.show()
```
