```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.datasets import load_iris
iris_dataset=load_iris()
X,y=iris_dataset.data,iris_dataset.target
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
l=Lasso().fit(X_train,y_train)
print("l.coef_:{}",format(l.coef_))
print("l.intercept_:{}",format(l.intercept_))
print("training score:{:.2f}",format(l.score(X_train,y_train)))
print("testing score:{:.2f}",format(l.score(X_test,y_test)))
r=Ridge().fit(X_train,y_train)
print("r.coef_:{}",format(r.coef_))
print("r.intercept_:{}",format(r.intercept_))
print("training score:{:.2f}",format(r.score(X_train,y_train)))
print("testing score:{:.2f}",format(r.score(X_test,y_test)))
import matplotlib.pyplot as plt
feature_names=iris_dataset.feature_names
plt.figure(figsize=(10, 5))
plt.plot(l.coef_, label='Lasso', marker='o')
plt.plot(r.coef_, label='Ridge', marker='x')
plt.xticks(range(len(feature_names)), feature_names, rotation=45)
plt.ylabel('Coefficient value')
plt.title('Lasso vs Ridge Coefficients')
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```
